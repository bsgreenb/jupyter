# Combos
<a id="combo-def"></a>

$$ {n \choose k} = \frac{n!}{k!(n-k)!} $$

# Pascal's Identity

$$ {n \choose k} = {n - 1 \choose k - 1} + {n - 1 \choose k} $$

## Proof


$$
\begin{align}
{n - 1 \choose k - 1} + {n - 1 \choose k} & = \frac{(n-1)!}{(k-1)!(n-k)!} +  \frac{(n-1)!}{k!(n-1-k)!} \\ 
& = {(n-1)!}(\frac{k}{k!(n-k)!} + \frac{n-k}{k!(n-k)!}) \\
& = {(n-1)!}(\frac{n}{k!(n-k)!}) \\ 
& = \frac{n!}{k!(n-k)!} = {n \choose k}  \quad\text {Definition of Combo}
\end{align}
$$

# Binomial Theorem

$$ (a+b)^n = \sum_{k=0}^{n}\binom{n}{k}a^{n-k}b^k $$

We know that this is true for n = 1, so we can prove it by induction for all positive numbers. So we seek to prove that

$$ 
\begin{align}
& (a+b)^{n+1} = \sum_{k=0}^{n+1}\binom{n+1}{k}a^{n+1-k}b^k \\
& (a + b)(a+b)^n \\
& (a + b)\sum_{k=0}^{n}\binom{n}{k}a^{n-k}b^k \quad\text {Induction} \\
& \left(\sum_{k=0}^{n}\binom{n}{k}a^{n-k + 1}b^k\right) + \left(\sum_{k=0}^{n}\binom{n}{k}a^{n-k}b^{k+1}\right) \\
& \left(\binom{n}{0}a^{n+1}b^0 + \sum_{k=1}^{n}\binom{n}{k}a^{n-k + 1}b^k\right) + \left(\sum_{k=1}^{n+1}\binom{n}{k-1}a^{n-k+1}b^{k}\right) \\
& \left(a^{n+1} + \sum_{k=1}^{n}\binom{n}{k}a^{n-k + 1}b^k\right) + \left(\binom{n}{n}a^{0}b^{n+1} + \sum_{k=1}^{n}\binom{n}{k-1}a^{n-k+1}b^{k}\right) \\
& a^{n+1} + b^{n+1} + \sum_{k=1}^{n}\binom{n}{k}a^{n-k + 1}b^k + \sum_{k=1}^{n}\binom{n}{k-1}a^{n-k+1}b^{k} \\
& a^{n+1} + b^{n+1} + \sum_{k=1}^{n}\binom{n+1}{k}a^{n-k + 1}b^k \quad\text {Pascal's Identity} \\
& \binom{n+1}{0} a^{n-0+1}b^0 + \binom{n+1}{n+1}a^{n-(n+1)+1}b^{n+1} + \sum_{k=1}^{n}\binom{n+1}{k}a^{n-k + 1}b^k \\
& \sum_{k=0}^{n+1}\binom{n+1}{k}a^{n-k + 1}b^k \\
\end{align}
$$ 


# Derivatives

## Definition as a function of x
$$ f'(x) = \lim_{\Delta x\to0} \frac{f(x + \Delta x) - f(x)}{\Delta x} $$
## Definition at a specific point
$$ f'(c) = \lim_{x\to c} \frac{f(x) - f(c)}{x - c} $$

### Differentiability at a specific point

Differentiability just means the limit in the definition of a derative exists at a point.

Continuity at a point means
$$ \lim_{x\to c}f(x) = f(c) $$

Continuity is necessary, but not sufficient for differentiability.  Differentiability at a point requires that the limit works from both left and right of c.

Proof that differentiability requires continuity.  Assume differentiability and:

$$ 
\begin{align}
& \lim_{x\to c}f(x) - f(c) \\
& \lim_{x\to c} \frac{(x-c)(f(x) -f(c))}{(x-c)} \\
& \lim_{x\to c} (x-c) \lim_{x\to c} \frac{(f(x) -f(c))}{(x-c)} \quad\text {Limit assumed to exist} \\
& (0)\lim_{x\to c} \frac{(f(x) -f(c))}{(x-c)} \\
& \lim_{x\to c}(f(x) - f(c)) = 0 \\ 
& \lim_{x\to c}f(x) - \lim_{x\to c}f(c) = 0 \\
& \lim_{x\to c}f(x) - f(c) = 0 \\
& \lim_{x\to c}f(x) = f(c)
\end{align} 
$$


# Power Rule

$$ \frac{d}{dx}[x^n] = nx^{n-1} $$

## Proof

$$
\begin{align}
& \frac{d}{dx}[x^n] \\
& \lim_{\Delta x\to0} \frac{(x + \Delta x)^n - x^n}{\Delta x}  \quad\text {Definition of a Limit} \\
& \lim_{\Delta x\to0} \frac{\sum\limits_{k=0}^{n}\binom{n}{k}x^{n-k}\Delta x^k) - x^n}{\Delta x}  \quad\text {Binomial Theorem} \\
& \lim_{\Delta x\to0} \frac{\binom{n}{0}x^n \Delta x^0 + \sum\limits_{k=1}^{n}\binom{n}{k}x^{n-k}\Delta x^k) - x^n}{\Delta x} \\
& \lim_{\Delta x\to0} \frac{\sum\limits_{k=1}^{n}\binom{n}{k}x^{n-k}\Delta x^k)}{\Delta x} \\
& \lim_{\Delta x\to0} {\sum\limits_{k=1}^{n}\binom{n}{k}x^{n-k}\Delta x^{k-1})} \\
& \lim_{\Delta x\to0} \binom{n}{1} x^{n-1}\Delta x^{1-1} + (0) \quad\text {delta x is 0 factor for k>1} \\
& \lim_{\Delta x\to0} nx^{n-1} = nx^{n-1} \\
\end{align} 
$$
